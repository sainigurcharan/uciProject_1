```python
#Import Dependencies
import json
import requests
import pandas as pd
import numpy as np
import scipy as sp 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
base_url = "https://api.exchangeratesapi.io/history?"

years = [2017, 2016, 2015, 2014, 2013, 2012, 2011, 2010, 2009, 2008, 2007]

results = []

for year in years:
    query_url = f"{base_url}start_at={year}-01-01&end_at={year}-12-31&base=USD"
    response = requests.get(query_url)
    response_json = response.json()
    results.append(response_json)
    print(response.status_code)
    
    print(query_url)
```

```python
#Results Data

results
```

```python
#Here we made a for loop that grabs the data in the key of "rates"
#Then we find the mean of the years worth of data
#Lastly we then made it into one DataFrame

df_list = pd.DataFrame()
for i in range(0, len(years)):
    rates = pd.DataFrame(results[i]['rates'])
 

    df_mean = rates.mean(axis=1)

    df_list[years[i]] = df_mean
    
df_list

df_list.index.name = "Currencies"
#reduce indeces

#to grab fro row use loc
```

```python
currency_list = df_list.loc[["USD", "EUR", "IQD", "PKR", "PHP", "SOS", "UAH", "COP", "GBP", "CNY", "GEL", "AUD", "HNL"], :]


currency_list


##List of the currencies to pull data from. 
#['Iraq',
#'Pakistan',
#'Philippines',
#'Somalia',
#'Ukraine',
#'Colombia',
#'United Kingdom',
#'United States',
#'China',
##'Georgia',
#'Australia',
#'Honduras']
```
